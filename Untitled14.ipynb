```python
class transformer(BaseEstimator, TransformerMixin):  # Corrected class name
    def __init__(self, cols):
        self.cols= cols
        self.label_encoders = {} # Use a dictionary to store label encoders for each column

    def fit(self, X, y=None):
        for col in self.cols:
            le = LabelEncoder() # Create a new LabelEncoder for each column
            X[col]= le.fit_transform(X[col])
            self.label_encoders[col] = le # Store the fitted encoder
        return self

    def transform(self, X):  # Correct indentation for transform method
        X_encoded= X.copy()
        for col in self.cols:
            le = self.label_encoders.get(col) # Retrieve the encoder for this column
            if col in self.label_encoders:
                X_encoded[col]= le.transform(X[col])
        return X_encoded


X= X_train.copy()
y= y_train.copy()

O1= transformer(Le_col)  # Use the corrected class name
O1.fit(X)
X= O1.transform(X)
```
